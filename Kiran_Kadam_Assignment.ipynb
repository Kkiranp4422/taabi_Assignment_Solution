```python
#Outlier Treatment
#  By IQR
Q1 = data2[numerical_columns].quantile(0.25)
Q3 = data2[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Identify outliers using the IQR method
Q1 = data2[numerical_columns].quantile(0.25)
Q3 = data2[numerical_columns].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
outliers = (data2[numerical_columns] < (Q1 - 1.5 * IQR)) | (data2[numerical_columns] > (Q3 + 1.5 * IQR))
print(outliers.sum())
```

```python
#  1st type of code for trating the outliers

# UL=(Q3 + 1.5 * IQR)
# LL=(Q3 - 1.5 * IQR)

# outlier_col=[numerical_columns]
# for i in range(len(outlier_col)):
#     UL=(Q3 + 1.5 * IQR)
#     LL=(Q3 - 1.5 * IQR)
#     a=data2[data[outlier_col[i]]>UL[i]]
#     data2.loc[a.index,outlier_col[i]]=UL[i]

#     b=data2[data2[outlier_col[i]]<LL[i]]
#     data2.loc[b.index,outlier_col[i]]=LL[i]
```

```python
#capping-- 2nd type of code for tratiing the outliers
df_capped = data2.copy()
for column in numerical_columns:
    lower_bound = Q1[column] - 1.5 * IQR[column]
    upper_bound = Q3[column] + 1.5 * IQR[column]
    df_capped[column] = np.where(data2[column] < lower_bound,
                                 lower_bound,
                                 np.where(data2[column] > upper_bound, upper_bound,
                                          data2[column]
                                          )
                                 )
```
